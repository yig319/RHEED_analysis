```python
from scipy.optimize import curve_fit

    
def trim_axes(axs, N):
    """
    Reduce *axs* to *N* Axes. All further Axes are removed from the figure.
    """
    axs = axs.flat
    for ax in axs[N:]:
        ax.remove()
    return axs[:N]

def show_grid_plots(xs, ys, labels=None, ys_fit=None, img_per_row=4, subplot_height=3, ylim=None):

    if type(labels) == type(None): labels = range(len(ys))

    fig, axes = plt.subplots(len(ys)//img_per_row+1*int(len(ys)%img_per_row>0), img_per_row, 
                             figsize=(16, subplot_height*len(ys)//img_per_row+1))    
    trim_axes(axes, len(ys))
    
    for i in range(len(ys)):
        
        if len(ys) <= img_per_row:
            index = i%img_per_row
        else:
            index = (i//img_per_row), i%img_per_row

        axes[index].title.set_text(labels[i])
        
        im = axes[index].plot(xs[i], ys[i], marker='.')
        
        if type(ys_fit) != type(None):
            im = axes[index].plot(xs[i], ys_fit[i])
        
        if type(ylim) != type(None):
            axes[index].set_ylim([ylim[0], ylim[1]])

        im.set_clim(0, 1) 
        
    fig.tight_layout()
    plt.show()
    

def show_grid_plots(xs, ys, labels=None, ys_fit=None, img_per_row=4, subplot_height=3, ylim=None):

    if type(labels) == type(None): labels = range(len(ys))

    fig, axes = plt.subplots(len(ys)//img_per_row+1*int(len(ys)%img_per_row>0), img_per_row, 
                             figsize=(16, subplot_height*len(ys)//img_per_row+1))    
    trim_axes(axes, len(ys))
    
    for i in range(len(ys)):
        
        if len(ys) <= img_per_row:
            index = i%img_per_row
        else:
            index = (i//img_per_row), i%img_per_row

        axes[index].title.set_text(labels[i])
        
        im = axes[index].plot(xs[i], ys[i], marker='.')
        
        if type(ys_fit) != type(None):
            im = axes[index].plot(xs[i], ys_fit[i])
        
        if type(ylim) != type(None):
            axes[index].set_ylim([ylim[0], ylim[1]])

    fig.tight_layout()
    plt.show()
       
def fit_exp(xs, ys, p_init, x_peaks=[], I_diff=1, ylim=None, camera_freq=500, add_previous=False):
    def exp_func(x, a, tau):
        return a*(1 - np.exp(-x/tau))
    
    tau_list, sign_list = [], []
    labels, ys_fit = [], []
    I_start_list, I_end_list = [], []
    
    for i in range(len(xs)):
        x = np.linspace(1e-5, 1, len(ys[i])) # use second as x axis unit
        
        # use I/I0, I0 is saturation intensity (last value) and scale to 0-1 based 
        n_avg = len(ys[i])//100+3
        if i != 0 and add_previous:
            I_start = I_end
        else:
            I_start = np.mean(ys[i][:n_avg])

            
        I_end = np.mean(ys[i][-n_avg:])
        
        if I_end - I_start == 0:
            ys[i] = (ys[i]-I_start)
        elif I_end < I_start:
            ys[i] = (ys[i]-I_start)/(-I_diff)
        else:
            ys[i] = (ys[i]-I_start)/I_diff # I-Imin/Imax-Imin (I/I0)
                    
        params, params_covariance = curve_fit(exp_func, x, ys[i], p0=p_init, bounds=[0.01, 1]) # curve_fit
        a, tau = params
        y_fit = exp_func(x, a, tau)
        sign_list.append('1-exp(-t/tau)')
        labels.append(f'index: {i+1};\nfitted: {np.round(a, 2)}*(1-exp(-t/{np.round(tau, 2)});\nI_diff={I_diff})')
        
        ys[i] = list(ys[i])
        tau_list.append(np.round(tau, 4))
        I_start_list.append(np.round(I_start, 4))
        I_end_list.append(np.round(I_end, 4))
        ys_fit.append(y_fit)

    return tau_list, labels, ys_fit, [sign_list, I_start_list, I_end_list]
```

```python
plt.rcParams.update(rc_plot)

# title = 'Treated substrate(step width=0.430.04m) - growth_1'
h5_para_file = '../../data/test6_gaussian_fit_parameters_all.h5'

# load data
sample_x, sample_y = load_curve(h5_para_file, 'growth_1.', 'spot_2', 'img_intensity', camera_freq=500, x_start=0, figsize=(12,4))

# detect peaks
x_peaks, xs, ys = detect_peaks(sample_x, sample_y, camera_freq=500, laser_freq=1, step_size=5, prominence=0.1)
plot_curve(sample_x, sample_y, x_peaks=x_peaks, plot_type='scatter', xlabel='Time (s)', ylabel='Intensity (a.u.)',
           figsize=(12, 4))

# fit exponential function
tau_list, labels, ys_fit, info = fit_exp(xs, ys, p_init=(1, 0.1), I_diff=8000, x_peaks=x_peaks, ylim=None, camera_freq=500)
sign_list, I_start_list, I_end_list = info
show_grid_plots(xs, ys, labels, ys_fit, ylim=None)
x_tau = np.linspace(0, len(tau_list)-1, len(tau_list))
plot_curve(x_tau, tau_list, plot_type='lineplot', xlabel='Laser ablation (count)', ylabel='Diffusion Time Constant (s)', 
           yaxis_style='linear', figsize=(12, 4))
```

```python
def analyze_curves(h5_para_file, growth_dict, spot, metric, camera_freq=500, title=None):
    
    tau_list_all, Imin_list_all, Imax_list_all, x_list_all = [], [], [], []
    for growth_name in list(growth_dict.keys()):
        # load data
        sample_x, sample_y = load_curve(h5_para_file, growth_name, 'spot_2', 'img_intensity', camera_freq=500, x_start=0, figsize=(12,4))

        # detect peaks
        x_peaks, xs, ys = detect_peaks(sample_x, sample_y, camera_freq=camera_freq, laser_freq=growth_dict[growth_name],
                                       step_size=5, prominence=0.1)
#         print(np.sum(x_peaks))

        # fit exponential function
        tau_list, labels, ys_fit, info = fit_exp(xs, ys, tau_=0.1, x_peaks=x_peaks, ylim=None, camera_freq=camera_freq)
        sign_list, I_start_list, I_end_list = info
        tau_list_all+=tau_list
        
        if x_list_all != []:
            x_list = x_end + np.linspace(0, len(tau_list)-1, len(tau_list)) / growth_dict[growth_name]
        else:
            x_list = np.linspace(0, len(tau_list)-1, len(tau_list)) / growth_dict[growth_name]
        x_end = x_list[-1]
            
        x_list_all.append(x_list)
        
    tau_list_all = remove_outlier(np.array(tau_list_all[2:]), 0.8, n=5)
    x_list_all = np.concatenate(x_list_all)[:len(tau_list_all)]
    
    return tau_list_all, x_list_all
```
